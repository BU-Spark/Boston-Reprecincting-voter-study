```python
import pandas as pd
import numpy as np
import dask.dataframe as dd
import requests
import re
import json
from numpy import cos, sin, arcsin, sqrt
from math import radians
from datetime import datetime
```

```python
MA_Block_Group = pd.read_csv("MA_Census_Block_group.csv")
MA_Block_Group
```

```python
MA_Block_Group = MA_Block_Group.rename(columns=MA_Block_Group.iloc[0])
MA_Block_Group
```

```python
MA_Block_Group = MA_Block_Group.iloc[1:]
MA_Block_Group
```

```python
MA_Block_Group_cols = list(MA_Block_Group.columns)
MA_Block_Group_cols
```

```python
Margin_cols_to_drop = []
for columns in MA_Block_Group_cols:
    if 'Margin' in columns:
        Margin_cols_to_drop.append(columns)
   

MA_Block_Group = MA_Block_Group.drop(columns = Margin_cols_to_drop)

MA_Block_Group
```

```python
MA_cols_v2 = list(MA_Block_Group.columns)

MA_cols_v2 = [re.compile(r"Estimate").sub("", m) for m in MA_cols_v2]

MA_cols_v2 = [re.compile(r"!!").sub("", m) for m in MA_cols_v2]
MA_cols_v2
```

```python
MA_Block_Group.columns = MA_cols_v2
MA_Block_Group
```

```python
BG_split_list = [x.split("Block Group ")[1] for x in MA_Block_Group['Geographic Area Name'].tolist()]

Block_group_list = [x[0] for x in BG_split_list]

MA_Block_Group['Block_Group'] = Block_group_list
MA_Block_Group
```

```python
CT_split_list = [x.split("Census Tract ")[1] for x in MA_Block_Group['Geographic Area Name'].tolist()]

CT_group_list = [x.split(',')[0] for x in CT_split_list]


MA_Block_Group['Census_Tract'] = CT_group_list
MA_Block_Group
```

```python
MA_Block_Group['Census_Tract_cleaned'] = MA_Block_Group['Census_Tract'].str.replace('.', '')
MA_Block_Group
```

```python
List_of_clean_formatted_tracts = []

for tracts in MA_Block_Group['Census_Tract']:
    if '.' not in tracts:
        clean_tracts = tracts+'.00'
        if len(tracts) < 4:
            clean_tracts = '0' + clean_tracts
            List_of_clean_formatted_tracts.append(clean_tracts)
        else:
            List_of_clean_formatted_tracts.append(clean_tracts)
        
    else:
        if len(tracts) < 4:
            clean_tracts = '0' + clean_tracts
            List_of_clean_formatted_tracts.append(clean_tracts)
        else:
            List_of_clean_formatted_tracts.append(clean_tracts)
            
List_of_clean_formatted_tracts
        #print(tracts)
```

```python
MA_Block_Group['Census_Tract_cleaned'] = List_of_clean_formatted_tracts
MA_Block_Group
```

```python
MA_Block_group_WP = pd.read_csv("MA_Block_groups_Ward_Precinct_merge.csv")
MA_Block_group_WP
```

```python
MA_Block_group_WP['TRACTCE10'] = MA_Block_group_WP['TRACTCE10'].astype(str)
```

```python
MA_Block_group_WP['Tract_formatted'] = MA_Block_group_WP['TRACTCE10'].str[:-2] + '.' + MA_Block_group_WP['TRACTCE10'].str[-2:]
MA_Block_group_WP


```

```python


MA_Block_group_demographics_WP = MA_Block_group_WP.merge(MA_Block_Group, left_on='Tract_formatted', right_on='Census_Tract_cleaned')
MA_Block_group_demographics_WP

```

```python
MA_Block_group_demographics_WP.to_csv("MA_Block_group_demographics_WP.csv", encoding = 'utf-8')
```
